```python
import pandas as pd
import numpy as np
import scipy.io.wavfile as wf
import os
import librosa
```

```python
f = "data/audios_development/3.wav"
y, sr = librosa.load(f, sr=22050)
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)  
mfcd=librosa.feature.delta(mfcc)
mfdd=librosa.feature.delta(mfcd)

print(mfdd.shape)
```

```python
ll = []
index = []
for f in os.listdir("data/audios_development"): 
    y, sr = librosa.load(f"data/audios_development/{f}", sr=22050)
    index.append(int(f.split(".")[0]))
    mfcc = librosa.feature.mfcc(y=y, sr=sr ,n_mfcc=13)
    mfcd=librosa.feature.delta(mfcc)
    mfdd=librosa.feature.delta(mfcd)
    ll.append(mfdd)
    print(f)
```

```python
df = pd.DataFrame(data=ll, index=index).sort_index()
```
